```python
# 코스피와 코스닥 전 종목의 주가 데이터를 수집해보자.
```

```python
import FinanceDataReader as fdr

# 한국거래소 종목 목록 가져오기
stock_list = fdr.StockListing("KRX")
# 코넥스 제외
stock_list = stock_list.loc[stock_list["Market"] != "KONEX"]
# 선물, ETF, ETN 등 비주식 제외
stock_list = stock_list.loc[stock_list["MarketId"] == "STK"]
# 우선주 제외 (종목명 끝이 '우'로 끝나는 것 제외)
stock_list = stock_list.loc[~stock_list["Name"].str.endswith('우')]

display(stock_list)
```

```python
# 전 종목 순회 및 데이터 저장
import time
import os

base_dir = os.path.dirname(os.getcwd())

for code, name in stock_list[['Code', 'Name']].values:
    print(code, name)
    while True:
        try:
            data = fdr.DataReader(code, "2015-01-01", "2025-01-01")
            if len(data) > 300:
                file_path = os.path.join(base_dir, "종목정보", f"{name}.csv") # 요즘 많이 쓰이는 f-string 사용.
                data.to_csv(file_path)
                # 길이가 300이상인 데이터를 name이라는 이름의 csv 파일로 저장한다. 상장한지 얼마 안 돼 데이터가 너무 작은 종목은 제외.
            time.sleep(1)
            break # 정상적으로 데이터 저장까지 완료되면 반복문에서 빠져나감.
        except:
            print("연결 끊어짐. 10분 대기.")
            time.sleep(10 * 60) # 연결이 끊어지면 10분 재움.
```
