**API(Application Programming Interface, 응용 프로그램 프로그래밍 인터페이스)란,<br>
데이터의 소유자가 자신의 데이터에 쉽게 접근할 수 있도록 프로그램과 데이터베이스를 연결해주는 매개체**<br>
최근에는 API를 더 쉽게 사용할 수 있게 도와주는 다양한 파이썬 패키지가 많으며, 퀀트 쪽에서도 크롤링→API로 변하는 추세임.<br>
<br>
FinanceDataReader: https://github.com/FinanceData/FinanceDataReader

```python
import FinanceDataReader as fdr
# StockListing 함수로 수집할 종목의 symbol을 얻고, 이 symbol을 DataReader의 입력으로 사용한다.

KP_stock_list = fdr.StockListing("KOSPI")
display(KP_stock_list)
```

```python
# 'KB금융(105560)' 주가 데이터
sp_data = fdr.DataReader("105560", "2024-03-23", "2025-03-23")
display(sp_data)
```

```python
SP500_stock_list = fdr.StockListing("SP500")
display(SP500_stock_list)
```

```python
sp_data = fdr.DataReader("MMM", "2024-03-23", "2025-03-23")
display(sp_data)
```

```python
# 코스피 지수, 다우존스 지수 데이터
KP_idx_data = fdr.DataReader("KS11", "2024-01-01", "2025-01-01")
display(KP_idx_data)

# Volume: 거래량. 이 날 거래된 총 주식의 수량
# Amount: 거래대금. 거래량 x 체결가격
# UpDown: 지수 등락 여부. 1=상승마감, 2=하락마감
# Comp: 변동 포인트(지수 차이). 전일 대비 몇 포인트 등락했는지 절대값
```

```python
DJI_idx_data = fdr.DataReader("DJI", "2024-01-01", "2025-01-01")
display(DJI_idx_data)
```

```python
# 기타 금융 데이터 수집도 가능하다. 원자재 선물, 환율, 채권 수익률 등.
display(fdr.DataReader("CL", "2024-01-01", "2025-01-01")) # WTI Crude Oil

# Adjusted Close Price: 수정 종가. 배당, 액면분할 및 병합 등의 이벤트를 반영한 실질적인 종가.
```
