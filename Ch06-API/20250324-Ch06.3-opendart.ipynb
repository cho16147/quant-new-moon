전자공시시스템 DART(Data Analysis, Retrieval and Transfer System) API를 활용하자.

```python
import OpenDartReader
import os
from dotenv import load_dotenv
# .env 파일 로드
load_dotenv()
api = os.getenv("API_KEY")
dart = OpenDartReader(api)

result = dart.list("105560", start = "2024-01-01", end = "2025-01-01")
display(result.head())
display(result.shape)

# corp_code : DART에 등록된 고유한 기업 코드
# stock_code: 주식 종목 코드
# corp_cls: classification. Y=상장, K=코스닥, N=비상장 등
# rcept_no: 해당 공시의 접수번호
# recept_dt: 접수일
# flr_nm: filer. 공시 제출인
# rm: remark. 비고, 특이사항 
```

```python
display(dart.report("105560", "배당", 2024, "11011"))
# report 메서드는 key_word, report_code를 인자로 받는다.
# key_word는 "증자", "배당", "임원", "타법인출자" 등등 다양함.
# "11013": 1분기보고서, "11012": 반기보고서, "11014": 3분가보고서, "11011": 사업보고서

# se: Section. 항목명 및 구분
# thstrm: This term. 당기
# frmtrm: Former Term. 전기
# lwfr: Last Year Former Term. 전년도 동기
# stlm_dt: 결산 기준일
```

```python
# finstate는 기업의 재무제표를 가져온다.
result = dart.finstate("105560", 2023, "11011")

# 이걸로 계정명을 확인할 수 있다. 특히 은행업은 한번쯤 확인해봄직함.
print(result["account_nm"].unique())

 # "fs_nm 칼럼이 "연결재무제표"인 행을 필터링
result = result.loc[result["fs_nm"]=="연결재무제표"]
# "account_nm" 칼럼이 "당기순이익", "영업이익", "매출액" 중 하나인 행을 필터링.
result = result.loc[result["account_nm"].isin(["이자수익", "이자비용", "순이자손익", "영업이익", "당기순이익(손실)", "당기순이익"])] 
# 해당 칼럼을 필터링.
result = result.loc[:, ["account_nm", "fs_nm", "sj_nm", "thstrm_dt", "thstrm_amount"]]
display(result)
```

```python
print(result["thstrm_amount"].dtype) # 데이터 타입 확인. 문자열이다.

def remove_comma(value):
    return float(value.replace(",", ""))

display(result["thstrm_amount"].apply(remove_comma))
```
