```python
import pandas as pd

df1 = pd.DataFrame({"A":[1,2,3], "B":[4,5,6]})
df2 = pd.DataFrame({"A":[7,8], "B":[9,10]})
display(df1)
display(df2)
```

```python
# concat 함수는 둘 이상의 시리즈 혹은 데이터프레임을 행 혹은 열 방향으로 이어붙인다. concatenate.
# axis = 0은 행을 기준으로 연산을 진행. 즉 세로 방향, 각 열끼리 연산 진행. 
display(pd.concat([df1, df2], axis = 0, ignore_index = True))
display(pd.concat([df1, df2], axis = 0, ignore_index = False))
# 보면, axis = 0으로 묶어줬기 때문에, ignore_index도 행 이름에 관여하게 된다. 대개 True를 사용.
```

```python
# axis = 1은 열을 기준으로 연산을 진행. 즉 가로 방향, 각 행끼리 연산 진행.
display(pd.concat([df1, df2], axis = 1, ignore_index = True))
display(pd.concat([df1, df2], axis = 1, ignore_index = False))
# 보면 axis = 1로 묶어줬기 때문에, ignore_index도 열 이름에 관여하게 된다. 대개 False를 사용.
```

```python
# merge 함수는 두 개의 데이터프레임을 join한다. 
# join이란 두 데이터프레임의 공통 칼럼을 기준으로 병합하는 행위.
# 이때의 공통 칼럼을 join key라고 한다.

df1 = pd.DataFrame({"직원": ["김철수", "박철수", "김민수", "이민수"],
                    "부서": ["회계", "데이터분석", "데이터분석", "인사"]})
df2 = pd.DataFrame({"직원": ["김철수", "이민수", "김민수", "박철수"],
                    "입사년도": [2004, 2008, 2012, 2014]})
merged_df = pd.merge(df1, df2, on = "직원")  # on 인자에 "직원"을 입력해서, df1과 df2를 "직원" 칼럼 기준으로 병합한다.
display(merged_df)
```

```python
# how 인자는 join 방법을 설정한다.
df1 = pd.DataFrame({"key":[1,2,3], "x":[1,2,3]})
df2 = pd.DataFrame({"key":[2,3,4], "y":[2,3,4]})
display(df1)
display(df2)
```

```python
# inner: join 키에 대한 교집합을 기준으로 병합한다.
display(pd.merge(df1, df2, on = "key", how = "inner"))
```

```python
# left: df1의 join 키에 포함된 [1,2,3]에 대해서만 병합이 이루어짐.
display(pd.merge(df1, df2, on = "key", how = "left"))
# right: df2의 join 키에 포함된 [2,3,4]에 대해서만 병합이 이루어짐.
display(pd.merge(df1, df2, on = "key", how = "right"))
```

```python
# outer: join 키에 대한 합집합을 기준으로 병합한다.
display(pd.merge(df1, df2, on = "key", how = "outer"))
```
