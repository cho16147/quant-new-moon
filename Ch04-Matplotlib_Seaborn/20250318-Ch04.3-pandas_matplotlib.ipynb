```python
import seaborn as sns
from matplotlib import pyplot as plt
from matplotlib import rcParams

sns.set()
%matplotlib inline

rcParams["font.family"] = "Malgun Gothic"
rcParams["axes.unicode_minus"] = False
```

```python
import os
import pandas as pd

base_dir = os.path.dirname(os.getcwd())  # 또는 notebook_dir = os.getcwd() 도 가능
file_path = os.path.join(base_dir, "data", "판다스", "재무제표샘플_대원미디어.csv")
df = pd.read_csv(file_path, encoding="euc-kr")
display(df)
```

```python
# kind 인자는 어떤 그래프를 그릴지 결정함.
df.plot(kind = "line", x = "연도", y = ["영업이익", "당기순이익"],
        figsize = (6,4),
        xticks = [2016, 2017, 2018, 2019, 2020])
plt.show()
```

```python
df.plot(kind = "bar", x = "연도", y = ["영업이익", "당기순이익"], figsize = (6, 4))
plt.show()
```

```python
df.plot(kind = "barh", x = "연도", y = ["영업이익", "당기순이익"], figsize = (6, 4))
plt.show()
```

```python
df.plot(kind = "box", y = ["영업이익", "당기순이익"])
plt.show()
```

```python
df.plot(kind = "hist", y = "영업이익", color = "black")
plt.show()
```
