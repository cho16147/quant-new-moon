**파이썬에서 데이터를 시각화하는 패키지는 대부분 맷플롯립을 바탕으로 한다.**<br>
다른 시각화 패키지를 사용하더라도 맷플롯립을 사용해 그래프 관련 환경설정을 하는 경우가 많음.

```python
from matplotlib import pyplot as plt
plt.scatter([1,2,3], [4,5,6]) # x 좌표 목록과 y 좌표 목록을 순서대로 입력받아 산점도를 그린다.
plt.show()
```

```python
plt.scatter([1,2,3], [4,5,6])
plt.scatter([1.5, 2.5], [3.5, 5.5])
plt.show()
```

```python
%matplotlib inline 
# 이 주피터 매직 커맨드는 plot이 별도의 창이 아닌 셀에 나타나게 한다.
plt.scatter([1,2,3], [4,5,6])
plt.show()
```

```python
plt.scatter([3,4,5], [7,7,7])
plt.title("example")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
import matplotlib.font_manager
fpaths = matplotlib.font_manager.findSystemFonts()
font_names = []
for i in fpaths:
    f = matplotlib.font_manager.get_font(i)
    font_names.append(f.family_name)

print(font_names[:10])
```

```python
for fn in font_names:
    if 'malgun' in fn.lower():
        print(fn)
```

```python
from matplotlib import rcParams
rcParams["font.family"] = "Malgun Gothic"
rcParams["axes.unicode_minus"] = False
```

```python
plt.scatter([3,4,5], [1,4,5])
plt.title("예제") # 한글이 잘 표시된다.
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
# 축 범위는 xlim, ylim 함수로 설정한다.
plt.scatter([1,2,3], [4,5,6])
plt.xlim(0,4)
plt.ylim(3,7.5)
plt.show()
```

```python
plt.scatter([1,2,3], [4,5,6])
plt.xlim(left = 0.5)
plt.ylim(top = 7)
plt.show()
# 설정하지 않은 값은 데이터 범위에 맞게 자동으로 설정된다.
```

```python
# 눈금 값은 xticks, yticks 함수로 설정한다.
plt.scatter([1,2,3], [4,5,6])
plt.xticks([1,2,3], ["A","B","C"])
plt.show()
```

```python
# label 인자는 그래프의 라벨, legned 함수는 라벨의 목록인 범례.
plt.scatter([1,1,1], [1,2,3], label = "파랑, x=1")
plt.scatter([-1,-1,-1], [1,2,3], label = "x=-1")
plt.legend(loc = "upper center") # 위치 지정 안하면 임의의 적당한 자리에 생성됨.
plt.show()
```

```python
# 가로 5인치, 세로 3인치의 plot 생성하겠다는 것. 비율상으로 가로가 세로보다 약 1.67배 더 길게 설정된다.
plt.figure(figsize = (5,3))
plt.scatter([13,3,5], [4,5,6])
plt.show()
```

```python
import seaborn as sns
sns.set() # 기본값(style = "darkgrid")
plt.scatter([1,2,3], [4,5,6])
plt.show()
```
