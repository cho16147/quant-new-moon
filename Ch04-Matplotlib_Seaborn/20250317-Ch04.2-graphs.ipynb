```python
import seaborn as sns
from matplotlib import pyplot as plt
from matplotlib import rcParams
# rcParmas은 Matplotlib에서 그래프의 전역 설정(global 설정값) 을 바꿀 수 있는 딕셔너리임.
# rc = resource configuration 쯤?

sns.set()
%matplotlib inline

rcParams["font.family"] = "Malgun Gothic"
rcParams["axes.unicode_minus"] = False
```

```python
plt.plot([1,2,3], [4,1,2], label = "line1") # (1,4), (2,1), (3,2)를 잇는 선 그래프 그리기
plt.plot([1,2,3], [5,6,0], label = "line2")
plt.legend(loc = "upper right")
plt.show()
```

```python
# 선 색깔 바꿔주기
plt.plot([1,2,3], [4,1,2], color = "black", label = "line 1")
plt.plot([1,2,3], [5,6,0], color = "green", label = "line 2")
plt.legend(loc = "upper right")
plt.show()
```

```python
# 선 스타일, 두께 바꿔주기
plt.plot([1,2,3], [4,1,2], linestyle = "dashed", linewidth = 6, label = "line 1")
plt.plot([1,2,3], [5,6,0], linestyle = "dotted", linewidth = 2, label = "line 2")
plt.legend(loc = "upper right")
plt.show()
```

```python
# 데이터 포인트를 표시하는 마커(marker) 설정
plt.plot([1,2,3], [4,1,2], marker = "s", markersize = 4, label = "line 1")
plt.plot([1,2,3], [5,6,0], marker = "o", markersize = 8, label = "line 2")
plt.show()
```

```python
# 그래프가 아닌 산점도 그리기 
plt.scatter([1,2,3], [4,1,2], label = "scatter1")
plt.scatter([1,2,3], [5,6,0], label = "scatter2")
plt.legend(loc = "upper right")
plt.show()
```

```python
# 산점도의 마커 설정
plt.scatter([1,2,3], [4,1,2], marker = "*", s = 40, color = "red", label = "scatter 1")
plt.scatter([1,2,3], [5,6,0], marker = "o", s = 80, color = "green", label = "scatter 2")
plt.legend()
plt.show()

# 보니까 legend 안의 마커 크기도 같이 커지네
# 참고로 선 그래프와 산점도의 마커 사이즈는 10배 차이가 난다.
```

```python
import numpy as np
x = np.random.normal(0, 1, 10000) # N(0, 1^2)를 따르는 만 개의 데이터포인트 
y = np.random.normal(0, 1, 10000)
plt.scatter(x, y, color='red')
plt.show()
```

```python
# alpha는 투명도.
plt.scatter(x, y, color = "red", alpha = 0.05)
plt.show()
```

```python
# 막대 그래프는 막대의 중심 위치를 직접 입력하므로, 실수에 주의.
plt.bar(range(3), [10, 20, 30])
plt.show()
```

```python
# 눈금 값 설정
plt.bar(range(3), [10,20,30])
plt.xticks(range(3), ["A", "B", "C"])
plt.show()
```

```python
# 스타일 설정. width는 막대 두께, linewidth는 막대 테두리 두께.
plt.bar(range(3), [10,20,30], width = 0.5, color = "pink", linewidth = 25, edgecolor = "black")
plt.xticks(range(3), ["A", "B", "C"])
plt.show()

# 멋져요
```

```python
plt.bar(
    range(3), [10, 20, 30], width = 0.5, color=["red", "blue", "black"], linewidth = 3.5, edgecolor = "yellow"
)
# width와 linewidth는 하나의 값만 입력했으므로, 모든 막대 두께, 테두리 두께는 동일하게 설정됨.
plt.xticks(range(3), ["A", "B", "C"])
plt.show()
```

```python
# 박스플롯은 데이터의 통계량과 이상치를 한 눈에 볼 수 있다.
x = np.random.normal(10, 5, 100)
plt.boxplot(x)
plt.show()
```

```python
x1 = np.random.normal(10, 5, 1000)
x2 = np.random.normal(10, 10, 100)
plt.boxplot([x1, x2])
plt.show()
# 박스플롯은 데이터의 개수에는 영향을 받지 않는 시각화 방식이다. 개수보다는 분포 파악에 용이함
# 다만 분포 비교시 표본 크기의 차이는 알고 보는게 좋다.
```

```python
# 박스플롯 눈금 값 설정
x1 = np.random.normal(10, 5, 1000)
x2 = np.random.normal(10, 10, 100)
plt.boxplot([x1, x2])
plt.xticks([1,2], ["x1", "x2"])
# 좌표가 1인 곳의 눈금 값을 "x1"으로, 2인 곳의 눈금 값을 "x2"로 설정한다.
plt.show()
```

```python
# 히스토그램
x = np.random.normal(10, 2, 100)
plt.hist(x)
plt.show()
```

```python
# 히스토그램의 구간 개수가 너무 작으면 값의 차이가 큰 두 값도 같은 집단으로 간주된다.
plt.hist(x, bins = 5)
plt.show()
```
